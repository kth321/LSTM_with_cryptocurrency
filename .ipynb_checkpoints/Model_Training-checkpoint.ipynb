```python
import os
import math
import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *
from keras.callbacks import EarlyStopping

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

## 데이터 불러오기

```python
BTC = pd.read_csv('bitcoin_indicator.csv')
RPL = pd.read_csv('repl_indicator.csv')
ETH = pd.read_csv('etherium_indicator.csv')
time = pd.read_csv('time.csv')
```

## 데이터 shape 확인

```python
print('shape')
print('shape BTC: ', BTC.shape)
print('shape RPL: ', RPL.shape)
print('shape ETH: ', ETH.shape)
print('shape time: ', time.shape)
```

```python
time_step = 30
n_days = 15
```

## training / test set

```python
idx = int(BTC.shape[0] * 0.8)

train_set = BTC.iloc[:idx].values
test_set = BTC.iloc[idx:].values

print('train set size: ', train_set.shape[0])
print('test set size: ', test_set.shape[0])
```

## 정규화(Standard Scaler)

```python
ss = StandardScaler()

train_set_scaled = ss.fit_transform(train_set)
test_set_scaled = ss.transform(test_set)

X_train = []
y_train = []

for i in range(time_step, idx - n_days):
    X_train.append(train_set_scaled[i-time_step:i])
    y_train.append(train_set_scaled[i+n_days, 0])
    
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
model = Sequential()

model.add(LSTM(100, return_sequences=True, input_shape=(X_train.shape[-2:])))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.4))

model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=80, batch_size=32)
```
