# 데이터 획득
- [비트코인, 리플, 이더리움]

```python
tickers = ['BTC', 'XRP', 'ETH']
ohlcv = []

for ticker in tickers:
    ohlcv.append(pybithumb.get_ohlcv(ticker, interval='hour12'))
ohlcv
```

전처리를 일괄적으로 하기 위해 데이터의 개수가 가장 적은 리플에 맞게 다른 가상화폐들의 행 개수도 동일하게 맞춰준다

```python
remove1 = len(ohlcv[0]) - len(ohlcv[1])
remove3 = len(ohlcv[2]) - len(ohlcv[1])

remove1, remove3
```

```python
ohlcv[0] = ohlcv[0].drop(ohlcv[0].head(remove1).index)
ohlcv[2] = ohlcv[2].drop(ohlcv[2].head(remove3).index)

for ticker_data in ohlcv:
    ticker_data.drop(ticker_data.tail(1).index, inplace=True)
```

데이터 처리의 편의를 위해 time index를 하나의 열로 바꿔주고 index를 초기화해줌

```python
for ticker_data in ohlcv:
    ticker_data.reset_index(inplace=True)
```

```python
ohlcv[0].to_csv('bitcoin.csv', index=False)
ohlcv[1].to_csv('repl.csv', index=False)
ohlcv[2].to_csv('etherium.csv', index=False)
```
